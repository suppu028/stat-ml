```python
data = data.drop(["Country", "Status", "Year"], axis = 1)
```

```python
def figure(x , y = "Life expectancy "):
  Y = data[y]
  X = data[x]
  plt.xlabel(x)
  plt.ylabel(y) 
  plt.scatter(X,Y)
```
